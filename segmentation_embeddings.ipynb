```python
import pandas as pd

# splits = {'train': 'minor/train.json', 'test': 'minor/test.json'}
# df = pd.read_json("hf://datasets/mlcore/arxiv-classifier/" + splits["train"], lines=True)
df = pd.read_csv("arxiv-minor-dataset.txt")
```

```python
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
%pip install torch
%pip install wtpsplit
```

```python
from wtpsplit import SaT
sat_sm = SaT("sat-3l-sm")
```

```python
splits_list = []
for split in sat_sm.split(df['fulltext'].tolist()[:2], remove_whitespace_before_inference=True, do_paragraph_segmentation=True):
  splits_list.append(split)
```

```python
paper_i = 1
paragraph_j = 6
sentence_k = 1
print(len(splits_list[paper_i][paragraph_j]))
print(splits_list[paper_i][paragraph_j])
```

```python
paragraphs_splits = []
for paper in splits_list:
  paras = []
  #print("\nPAPER")
  for paragraph in paper:
    clean_sentences = [sentence.replace("\n", "") for sentence in paragraph]
    para = " ".join(clean_sentences)
    paras.append(para)
  paragraphs_splits.append(paras)
```

```python
model = SentenceTransformer("jordyvl/scibert_scivocab_uncased_sentence_transformer")
data = paragraphs_splits

split1_embeddings = model.encode(data[0])
split2_embeddings = model.encode(data[1])

print("Embeddings for First Split")
print(split1_embeddings)
print(split1_embeddings.shape)

print("Embeddings for Second Split")
print(split2_embeddings)
print(split2_embeddings.shape)
```

```python
import numpy as np

np.save("split1_embeddings.npy", split1_embeddings)
np.save("split2_embeddings.npy", split2_embeddings)

loaded_split1 = np.load("split1_embeddings.npy")
loaded_split2 = np.load("split2_embeddings.npy")

print("Loaded Split 1 Embeddings")
print(loaded_split1)
```
