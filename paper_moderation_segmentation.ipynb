```python
from wtpsplit import SaT
sat_sm = SaT("sat-3l-sm")
```

```python
splits_list = []
for split in sat_sm.split(df['fulltext'].tolist()[:2], do_paragraph_segmentation=True):
  splits_list.append(split)
```
